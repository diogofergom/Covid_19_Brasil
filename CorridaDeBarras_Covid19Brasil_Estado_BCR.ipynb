```python
import pandas as pd
import numpy as np
import seaborn as sns
from IPython.display import Image


import matplotlib.pyplot as plt
%matplotlib inline

import plotly_express as px
import bar_chart_race as bcr
```

# Lendo os arquivos

```python
df_covidbr2021_parte1 = pd.read_csv('HIST_PAINEL_COVIDBR_2021_Parte1_08dez2021.csv', sep=';')
```

```python
df_covidbr2021_parte2 = pd.read_csv('HIST_PAINEL_COVIDBR_2021_Parte2_08dez2021.csv',sep=';')
```

# Analise previa

```python
df_covidbr2021_parte1.info()
```

### Dados fantantes

```python
df_parte1 = df_covidbr2021_parte1.isna().sum()
df_parte1
```

```python
df_parte2 = df_covidbr2021_parte2.isna().sum()
df_parte2
```

## Concatenando os datasets

```python
df_covidbr = pd.concat([df_covidbr2021_parte1,df_covidbr2021_parte2])
```

```python
df_covidbr.isna().sum()
```

## Filtrar os dados pela coluna 'codmun'

```python
# Transformando dados NaN em zero
df_covidbr.fillna(0,inplace=True)
df_covidbr
```

```python
df_covidbr = df_covidbr[df_covidbr['codmun'] == 0]
```

```python
df_covidbr.shape
```

### Quantidade de Brasil na coluna estado

```python
df_covidBrasil = df_covidbr[df_covidbr['regiao'] == 'Brasil']
df_covidBrasil.head()
```

```python
df_covidBrasil.info()
```

## Criando o DataFrame somente com os estados

```python
df_CovidEstados = df_covidbr.drop(df_covidbr.loc[df_covidbr['regiao'] == 'Brasil'].index)
```

```python
df_CovidEstados
```

### Gráfico de Corrida de Barras (Bar Chart Race)

### Tratando os dados

```python
df_Covid19Estados = df_CovidEstados.groupby(['estado','data'])['casosNovos'].mean()
```

```python
df_Covid19Estados
```

### Transformando em DataFrame

```python

df_Covid19Estados = pd.DataFrame(df_Covid19Estados).unstack().T
```

```python
df_Covid19Estados
```

```python
# O droplevel = apagar o casosNovos do multIndex
df_Covid19Estados = df_Covid19Estados.droplevel(level=0)
```

```python
df_Covid19Estados
```

#### df_Covid19Estados com dados agregados

```python
df_Covid19Estados = df_Covid19Estados.cumsum()
```

```python
df_Covid19Estados.head()
```

### Aplicando o método

```python
#df = bcr.load_dataset('covid19_tutorial')
bcr.bar_chart_race(
    df=df_Covid19Estados,
    title='Casos Novos de Covid por estados - Brasil',
    orientation='h',
    sort='desc',
    n_bars=12,
    filename='CasosNovosCovidEstados2.mp4'
)
```
